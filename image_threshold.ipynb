```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
img=cv2.imread('/content/chrome-256.png')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
print("gray image is\n")
cv2_imshow(gray)
ret,thresh=cv2.threshold(gray,130,255,cv2.THRESH_BINARY)
print("binary image is\n")
cv2_imshow(thresh)
print(thresh.shape)
```
